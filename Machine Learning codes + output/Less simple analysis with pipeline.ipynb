```python
import pandas as pd
import random as rd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.svm import LinearSVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
```

```python
all_tweets = pd.read_csv('all_tweets_data.csv')
all_tweets
```

```python
twters = all_tweets['user'].unique()
twt_user_1 = all_tweets[(all_tweets.user == twters[0]) & (all_tweets['Buy/Neutral/Sell'] != 0)]
twt_user_2 = all_tweets[(all_tweets.user == twters[1]) & (all_tweets['Buy/Neutral/Sell'] != 0)]
twt_user_3 = all_tweets[(all_tweets.user == twters[2]) & (all_tweets['Buy/Neutral/Sell'] != 0)]
twt_user_4 = all_tweets[(all_tweets.user == twters[3]) & (all_tweets['Buy/Neutral/Sell'] != 0)]
twt_user_5 = all_tweets[(all_tweets.user == twters[4]) & (all_tweets['Buy/Neutral/Sell'] != 0)]
```

```python
twt_user_2
```

```python
vect = CountVectorizer()
lr = LogisticRegression()

pipe_lr = Pipeline([
    ('vect', vect),
    ('lr', lr)
])

svm = LinearSVC()

pipe_svm = Pipeline([
    ('vect',vect),
    ('svm',svm)
])

nb = MultinomialNB()

pipe_svm = Pipeline([
    ('vect',vect),
    ('nb',nb)
])
```

```python
pipe_svm.steps
```

```python
X = twt_user_2.new_tweets
y = twt_user_2['Buy/Neutral/Sell']
#cross_val_score(pipe, X, y, cv=5, scoring='accuracy')
```

```python
param_grid_lr={
    'vect__stop_words':[None, 'english'],
    'vect__ngram_range': [(1,1), (1,2), (1,3)],
    'lr__C': [100,10,1,0.1,0.01]
}

param_grid_svm={
    'vect__stop_words':[None, 'english'],
    'vect__ngram_range': [(1,1), (1,2), (1,3)],
    'svm__C': [100,10,1,0.1,0.01]
}

param_grid_nb={
    'vect__stop_words':[None, 'english'],
    'vect__ngram_range': [(1,1), (1,2), (1,3)],
    'nb__C': [100,10,1,0.1,0.01]
}

grid_lr = GridSearchCV(pipe_lr, param_grid = param_grid_lr, cv=5, scoring='accuracy')

grid_svm = GridSearchCV(pipe_svm, param_grid = param_grid_svm, cv=5, scoring='accuracy')

grid_nb = GridSearchCV(pipe_nb, param_grid = param_grid_svm, cv=5, scoring='accuracy')
```

```python
grid_lr.fit(X, y)
grid_svm.fit(X, y)
```

```python
print(grid_lr.best_params_)
print(grid_lr.best_score_)
```
